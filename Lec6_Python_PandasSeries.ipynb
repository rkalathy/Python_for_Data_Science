Pandas has two datastructures: Series and Dataframes

```python
# Import pandas

import pandas as pd
```

```python
# Creating series from a list

s1 = pd.Series([10,20,30,40,50])
print(s1)
```

```python
type(s1) #checking the type of s1 series created
```

```python
# creating an empty series

s2 = pd.Series()
print(s2)
```

```python
# create a series from numpy array

import numpy as np
arr = np.array([4,2,6,8,9,10])
serie = pd.Series(arr)
print(serie)
```

```python
# create a pandas series from a dictionary:

data = {"A":10,"B":40,"X":60,"Y":90,"Z":100}
s4 = pd.Series(data)
print(s4)
```

```python
print(s4.iloc[0]) #iloc is used to access certain index location
```

```python
print(s4.loc["B"]) #loc is used to access value based on label index
```

```python
s4.iloc[0:4]
```

```python
s4.loc["A":"Y"]
```

### Filter or subset

```python
h1 = pd.Series([3,4,5,6,7,8,10,1,2,3,4,5,667,7,88,8,99,9])
print(h1)
```

```python
save = h1[h1>=8]
save
```

```python
save2 = h1[(h1>=8) & (h1<=500)]
save2
```

```python
type(save2)
```

```python
new = save2.reset_index(drop=True)
new
```
