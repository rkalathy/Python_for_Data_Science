```python
# Bar chart - 2nd option

plt.bar(["City A","City B","City C"],[90,50,110])
plt.show()
```

```python
x = [40,20,5]
labels_vehicles = ["Bikes","Cars","Buses"]

plt.pie(x,labels=labels_vehicles,autopct="%1.1f%%")
plt.show()
```

```python
# Data
labels = ["Car", "Bus", "Bike", "Train"]
sizes = [40, 25, 20, 15]

# Explode only the "Bus" slice
explode = [0, 0.3, 0, 0]   # 0.2 pushes Bus slice outward

# Define a custom color palette
colors = ["#FFB6B9", "#FAE3D9", "#BBDED6", "#8AC6D1"] 


# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(
    sizes,
    labels=labels,
    explode=explode,
    autopct="%1.1f%%",   # show percentages
    shadow=True,   # add shadow effect
    colors=colors
)

plt.title("Transport Mode Distribution")
plt.show()
```

```python
aa = np.arange(10)

plt.plot(aa,label = "Simple",color = "magenta")
plt.plot(aa+5, label = "Plus 5",color = "cyan")
plt.plot(aa+10,label = "Plus 10",color = "pink")

plt.legend()
plt.title("My mathematical plots")
plt.show()
```

```python
aa = np.arange(10)

plt.plot(aa,"-.",label = "Simple",color = "magenta")
plt.plot(aa+5,"--",label = "Plus 5",color = "cyan")
plt.plot(aa+10,":",label = "Plus 10",color = "red")

plt.legend()
plt.title("My mathematical plots")
plt.show()
```

```python
aa = np.arange(10)

plt.plot(aa,"-.",label = "Simple",color = "magenta",marker = "o")
plt.plot(aa+5,"--",label = "Plus 5",color = "cyan",marker="*")
plt.plot(aa+10,":",label = "Plus 10",color = "red",marker="d")

plt.legend()
plt.title("My mathematical plots")
plt.show()
```

```python
y = np.arange(1,20)
plt.plot(y,"D",y*10,"^",y*20,"s")

plt.title("Title of the graph")
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.grid(True)
plt.show()
```

# Example with Dataset and Visualization in Seaborn

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
# barplot
# Fare by Class

sns.barplot(x="class",y="fare",data=titanic,hue="class")
plt.title("Fare by Class")
plt.show()
```

```python
# barplot

sns.barplot(x="sex",y="survived",data=titanic,hue="class",palette="pastel")
plt.title("Survival by Gender")
plt.show()
```

```python
# barplot

sns.barplot(x="class",y="survived",data=titanic,hue="class",palette="pastel")
plt.title("Survival by Class")
plt.show()
```

```python
# countplot


sns.countplot(x="survived",data=titanic,palette="Set2",hue="survived")
plt.title("Survival count")
plt.show()
```

```python
# histogram

sns.histplot(titanic["age"].dropna(),color="pink")
plt.title("Age distribution of passengers")
plt.show()
```

```python
# boxplot

sns.boxplot(x="survived",y="fare",data=titanic,hue="sex",palette="Set3")
plt.show()
```

```python
# scatterplot

sns.scatterplot(x="fare",y="age",data=titanic,color="Green")
plt.show()
```

```python
# scatterplot

sns.jointplot(x="fare",y="age",data=titanic,color="red")
plt.show()
```

```python
sns.histplot(x="fare",data=titanic,color="purple",kde="True")
plt.show()
```

```python
# boxplot

sns.boxplot(x="class",y="age",data=titanic,hue="class",palette="rainbow")
plt.show()
```

```python
# countplot


sns.countplot(x="class",data=titanic,palette="Set2",hue="survived")
plt.title("Survival count")
plt.show()
```

```python
subset = titanic[["age","fare"]]
subset
```

```python
subset.corr()
```

# Plotly - Interactive Plots

```python
import plotly.express as px
```

```python
fig = px.box(titanic,x="survived",y="fare",color="survived")

# Change figure size
fig.update_layout(
    width=600,   # width in pixels
    height=500   # height in pixels
)
fig.show()
```
