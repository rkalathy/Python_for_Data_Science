## Indexing in numpy array - starts with 0

##### Syntax: array[start:end index(exclusive):step(Optional)]

```python
print(c2[1]) #Access element in the first index location
```

```python
print(c2[2:5]) #Access element from 2nd index until 5th index but excluding the 5th index
```

### Slicing in numpy arrays:

Syntax: array[Rows that you want,Columns that you want]

Row Format: [start:end:step] <br>
Column format: [start:end:step]

```python
nn1 = np.array([[2,5,6,7,8],[2,9,1,3,4]])
nn1
```

```python
nn1[0:2,2:4] #array of rows,column. Return 0th and 1st Row, 2nd and 3rd columns
```

```python
nn1[:,2:4] #Another way of writing/slicing
```

```python
nn1[:,2:6] #another example
```

### View and Copy in numpy arrays.

###### View will sahre the memory with original array, Copy creates a separate memory block

```python
nn1_copy1 = nn1
print(nn1_copy1)
```

```python
nn1_copy2 = np.copy(nn1)
print(nn1_copy2)
```

```python
nn1 = np.arange(10,15)
print(nn1)
```

```python
nn1_copy1
```

```python
nn1_copy2
```

```python
v = nn1.view()
print(v)
```

```python
print(v[0]) #access first element of the view
```

```python
v[0] = 999 # Changing the element in the view to check if changes reflected in original or not
print(v)
```

```python
nn1 #Original array also changed with changes made in view because they share memory
```

## Vector Operations in Python

#### Numpy is very good at mathematical operations and element wise operations

```python
v1 = np.array([1,3,5])
v2 = np.array([12,32,56])
```

```python
# Addition - element wise:

sum_vector = v1+v2
print(sum_vector)
```

```python
# substraction

sub_vector = v2-v1
print(sub_vector)
```

```python
# multiplication

mult_vector = v2*v1
print(mult_vector)
```

```python
# Multiplication by scaler
c = 10

vec = c*v1
print(vec)
```

```python
# Addition of a scaler
c = 10

vec2 = c+v1
print(vec2)
```

```python
#comparison

v1>v2
```

```python
# dot product of matrices

matrix1 = np.array([[2,3],[4,5]])
matrix2 = np.array([[4,5],[1,1]])

matmul1 = np.dot(matrix1,matrix2) #Option 1

matmul2 = matrix1 @ matrix2 #Option 2
```

```python
print(matmul1)
```

```python
print(matmul2)
```

```python
print(np.linalg.norm(v1)) #magnitude of vector
```
